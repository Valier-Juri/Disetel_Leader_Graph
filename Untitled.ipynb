```python
def generate_coordinates(n,p,q):
    coordinates = []
    for i in range(n+1):
        for j in range(p**i):
            for k in range(q**(n-i)):
                coordinates.append((i, j, k))
    return coordinates
```

```python
def transform_coordinates(coords, n, p, q):
    transformed_coords = []
    edge_library = []
    for coord in coords:
        i, j, k = coord
        transformed_coord = (i/n, (2*j+1)/(2*p**i), (2*k+1)/(2*q**(n-i)))
        transformed_coords.append(transformed_coord)
        if i < n:
            childrens = [(i+1, j*p+x,int(k/q)) for x in range(p)]
            transformed_childrens = [ (i/n, (2*j+1)/(2*p**i), (2*k+1)/(2*q**(n-i))) for (i,j,k) in childrens]
            for children in transformed_childrens:
                edge_library.append((transformed_coord,children))
    return transformed_coords, edge_library
```

```python
coords = generate_coordinates(1, 2, 2)
transformed_coords , edge_library = transform_coordinates(coords, 1, 2, 2)
print(transformed_coords , edge_library)
len(edge_library)
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

n = 5
p = 2
q = 2

coords = generate_coordinates(n, p, q)
print("5%")
transformed_coords, edge_library = transform_coordinates(coords, n, p, q)
print("10%")
z, x, y = zip(*transformed_coords)
print("20%")
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z,c="red")

ax.set_xlabel('Transformed First Entry')
ax.set_ylabel('Transformed Second Entry')
ax.set_zlabel('Transformed Third Entry')
print("40%")
# Plot the edges
for edge in edge_library:
    z_edge = [edge[0][0], edge[1][0]]
    x_edge = [edge[0][1], edge[1][1]]
    y_edge = [edge[0][2], edge[1][2]]
    ax.plot3D(x_edge, y_edge, z_edge, 'b')

plt.show()
```
